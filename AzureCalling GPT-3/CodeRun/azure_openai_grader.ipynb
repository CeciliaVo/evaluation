```python
#Note: The openai-python library support for Azure OpenAI is in preview.
#Note: This code sample requires OpenAI Python library version 1.0.0 or higher.
import os
from openai import AzureOpenAI

llm = AzureOpenAI(
  azure_endpoint = "https://autograding-testing.openai.azure.com/",
  api_key=os.environ["AZURE_OPENAI_KEY"],
  api_version="2024-02-15-preview"
)
```

```python
import os

path = 'C:/Users/Admin/Desktop/GPT4-INFO/AzureCalling GPT-3/CodeRun/demo'
os.chdir(path)

def get_students_work(file_path):
    with open(file_path, 'r') as file:
        students_work = file.read()
        
    return students_work
```

```python
import os

path = 'C:/Users/Admin/Desktop/GPT4-INFO/AzureCalling GPT-3/CodeRun/crit'
os.chdir(path)

def get_criteria(file_path1):
    with open("MarkingCriteria.txt", 'r') as file:
        criteria = file.read()
    with open("question1.txt", 'r') as file:
        question = file.read()
    return criteria, question
```

```python
def score_answer(criteria, question, students_work):
    # Convert the tuple into a string
    prompt = "I want you to act as an AI evaluating on students' coding work. I will provide you a text file of the coding question with points of the question in brackets (example: Question 1 (8pts)) and a text file of the standard marking criteria that you will need to follow to give recommended marks. I will provide you a folder contains coding works in text format and your task is to use artificial intelligence tools, such as natural language processing, to detect the programming language, give tailored feedback on how they can improve and write it in the Feedback section, only give 1 total recommended marks of the question point in the Score section, separated by a break line. Only use the marking criteria for standardizing the mark, do not give mark completely based on the provided marking criteria. You will also be given a sample marking from real instructors for each question in an Excel file, please also use this Excel file to give out the best recommended mark and tailored feedback for students. Here are the marking criteria, test question and the students' work in text: ".join([criteria, question, students_work])
    # print("prompt",prompt)
    
    messages = [
        {"role": "system", "content": "You are an excellent university lecturer of IT and software engineering programs."},
        {"role": "user", "content": prompt}
    ]
    # print("messages",messages)
    completion = llm.chat.completions.create(
        model="AzureCodeGrader", # model = "deployment_name"
        messages=messages,
        temperature=0,
        max_tokens=800,
        top_p=0.95,
        frequency_penalty=0,
        presence_penalty=0,
        stop=None
    )

    return completion.choices[0].message.content
```

```python
import os


# Get a list of all files in the 'StudentWork' directory
file_list = os.listdir('C:/Users/Admin/Desktop/GPT4-INFO/AzureCalling GPT-3/CodeRun/demo')

# Loop through each file in the directory
for file_name in file_list:
    file_path = os.path.join('C:/Users/Admin/Desktop/GPT4-INFO/AzureCalling GPT-3/CodeRun/demo', file_name)


    # Call the function to get user inputs
    students_work = get_students_work(file_path)
    criteria, question = get_criteria(file_path)
    # question = get_question(file_path)

    # Now you can use these inputs in your score_answer function
    score = score_answer(question, students_work, criteria)

    print(f"This is the AI-generated evaluation for {file_name}: \n{score}", end='\n')
```
