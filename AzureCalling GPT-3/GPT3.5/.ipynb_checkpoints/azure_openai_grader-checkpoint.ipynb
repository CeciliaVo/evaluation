```python
#Note: The openai-python library support for Azure OpenAI is in preview.
#Note: This code sample requires OpenAI Python library version 1.0.0 or higher.
import os
from openai import AzureOpenAI

llm = AzureOpenAI(
  azure_endpoint = "https://autograding-testing.openai.azure.com/",
  api_key=os.environ["AZURE_OPENAI_KEY"],
  api_version="2024-02-15-preview"
)
```

```python
def score_answer(instruction, student_answer, marking_criteria):
    # Convert the tuple into a string
    prompt = ' '.join( [instruction, student_answer, marking_criteria])
    print("prompt",prompt)
    messages = [
        {"role": "system", "content": "You are a teacher assistant who help to evaluate and give mark student code assesment based on the marking criteria."},
        {"role": "user", "content": prompt}
    ]
    print("messages",messages)
    completion = llm.chat.completions.create(
        model="AzureGrader", # model = "deployment_name"
        messages=messages,
        temperature=0.7,
        max_tokens=800,
        top_p=0.95,
        frequency_penalty=0,
        presence_penalty=0,
        stop=None
    )

    return completion.choices[0].message.content
```

```python
import os

def get_user_inputs(file_path):
    with open(file_path, 'r') as file:
        student_answer = file.read()

    with open("Test Questions.txt", 'r') as file:
        instruction = file.read()

    with open("Marking Criteria.txt", 'r') as file:
        marking_criteria = file.read()

    return instruction, student_answer, marking_criteria

# Get a list of all files in the 'StudentWork' directory
file_list = os.listdir('demo')

# Loop through each file in the directory
for file_name in file_list:
    file_path = os.path.join('demo', file_name)

    # Call the function to get user inputs
    instruction, student_answer, marking_criteria = get_user_inputs(file_path)

    # Now you can use these inputs in your score_answer function
    score = score_answer(instruction, student_answer, marking_criteria)

    print(f"Score for {file_name}: {score}")
```
